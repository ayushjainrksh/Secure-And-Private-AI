```python
#Loading image data
#Importing the packages
import torch
from torch import nn
import torch.nn.functional as F
from torchvision import transforms, datasets
import numpy as np
import matplotlib.pyplot as plt
```

[Download Dataset](https://s3.amazonaws.com/content.udacity-data.com/nd089/Cat_Dog_data.zip)

```python
#Import dataset and apply image augmentaion
train_transform = transforms.Compose([transforms.RandomRotation(30),
                               transforms.RandomResizedCrop(224),
                               transforms.RandomHorizontalFlip(),
                               transforms.ToTensor(),
                               transforms.Normalize((0.5,),(0.5,))])

test_transform = transforms.Compose([transforms.Resize(255),
                                    transforms.CenterCrop(224),
                                    transforms.ToTensor(),
                                    transforms.Normalize((0.5,),(0.5,))])

trainset = datasets.ImageFolder('Cat_Dog_data/train', transform=train_transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=32)

testset = datasets.ImageFolder('Cat_Dog_data/test', transform=test_transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=32)
```
